# Throughput Bench: Google Colab Quick Start

A small CLI for measuring inference throughput (img/s) of vision backbones on a single GPU, with a focus on geospatial workloads. Covers 33 timm architectures (ResNet, EfficientNet, ConvNeXt, MobileNet, RegNet, ViT, DeiT3, Swin, BEiT, DinoV3, CoAtNet) plus 12 geospatial foundation-model encoders (DOFA, CROMA, SenPaMAE, Galileo, OlmoEarth) under fp32 / fp16 / bf16 / AMP and `torch.compile`. Results are appended to a per-GPU CSV; an interactive webapp turns those numbers into "how fast can each backbone map the world?"

**Official repo:** https://github.com/calebrob6/throughput-bench

**Figure:** [Globe Race webapp](webapp/index.html) — pick two backbones and a GSD; the dot grid fills in proportional to land area each model has processed at its measured throughput.
